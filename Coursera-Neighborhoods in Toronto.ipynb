## Section 1: Create a dataframe of Toronto postal codes, bouroughs, and neighborhoods information

This section will obtain the postal codes, bourough, and neighborhoods data for the city of Toronto by scraping the Wikipedia webpage titled ['List of Postal Codes of Canada: M'](https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M).

The scraped data will then be processed to create a dataframe consisting of 3 columns:
1. Postal Code
1. Bourough name
1. Neighborhoods within the bourough

```python
from bs4 import BeautifulSoup as Soup #library for webscraping
import numpy as np #library for vectorized computation
import pandas as pd #library to process data as dataframes
import requests #library to make HTTP requests
```

This section of code accesses the webpage URL and scrapes the contents into a BeautifulSoup object. The table of pertinent information is then extracted into a variable. From the webpage, it is clear that the first table contains the necessary data. 

```python
#Define the website URL for the Toronto neighborhoods data
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

#Get the data and create a BeautifulSoup object of the Toronto wiki site
url_resp = requests.get(url)
url_soup = Soup(url_resp.text)

#Get the first table from the Toronto wiki site that includes the postal codes, bouroughs, and neighborhoods
table = url_soup.find('table')
```

Each row of the table is accessed and the postal code, bourough, and neighborhoods data is extracted into a list. The list is finally converted into a Pandas dataframe.

Key assumptions
- Any row with a bourough listed as 'Not assigned' is ignored
- The postal code is indicated with a 'p' tag 
- The postal code is only 3 characters in length
- The bourough and neighborhoods text are indicated with a 'span' tag
- The bourough is the first text in the 'span' tag
- Neighborhoods are text enclosed in parentheses

```python
table_contents=[]
#iterate through each row of the table and parse out the postal code, bourough, and neighborhoods information
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned': #skip table rows that do not contain useful information
        pass
    else:
        cell['PostalCode'] = row.p.text[:3] #get only the first 3 text characters for the postal code
        cell['Borough'] = (row.span.text).split('(')[0] #bourough is the first text before the parentheses
        
        #get Neighborhoods and eliminate any parentheses. Separate multiple neighborhoods by commas instead of slashes
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

#convert data into a Pandas dataframe
tor_df=pd.DataFrame(table_contents)

#go through the dataframe and rename bouroughs appropriately
tor_df['Borough']=tor_df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})

#increase the max column width displayed for a dataframe so it is easier to see Neighborhoods
pd.set_option('max_colwidth', 100)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

tor_df
```

## Section 2: Adding latitude and longitude to the Toronto dataframe

This section will expand the Toronto dataframe (tor_df) to include 2 new columns: Latitude and Longitude. These columns will hold the latitude and longitude geographical coordinates for each borough. 

The latitude and longitude coordinates are accessed from this [CSV file](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv).


```python
# import geocoder library to get latitude/longitude data
# this is commented out because geocoder is not used
#import geocoder 

# read in a CSV file containing geographical coordinates for each postal code and convert it to a Pandas dataframe
geo_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv'
geo_df = pd.read_csv(geo_url)

geo_df.head() # print out first 5 rows of geo_df to see what the dataframe looks like
```

This creates a function called get_latlong that returns the geographical latitude and longitude for a given postal code

- Input: postal_code should be type string
- Returns: latitude, longitude type float

```python
# create a function that takes in a postal code and returns the latitude and longitude
def get_latlong (postal_code):
    # initialize your variable to None
    #lat_lng_coords = None
    
    # loop until you get the coordinates
    # Ignore geocoder code because it is taking too long to query the data
    #while(lat_lng_coords is None):
        #g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
        #lat_lng_coords = g.latlng
    #latitude = lat_lng_coords[0]
    #longitude = lat_lng_coords[1]
   
    # find the row index for the matching Postal Code
    index = geo_df.loc[geo_df['Postal Code'] == postal_code].index[0]
    
    # get the latitude and longitude data from the geo_df dataframe using the index 
    latitude = geo_df['Latitude'][index]
    longitude = geo_df['Longitude'][index]
    
    return latitude, longitude
```

This section iterates through the tor_df rows and uses the postal code to retrieve the latitude and longitude from the function get_latlong. Finally, the tor_df dataframe is updated to include columns for the latitude and longitude data.

```python
# initialize empty lists to hold the latitude, longitude data
latitude = []
longitude = []

# iterate through each row of the Toronto dataframe and get the latitude and longitude for each postal code
for post_code in tor_df['PostalCode']:
    # pass the postal code for the current row to the get_latlong function to retrieve the latitude and longitude
    lat, long = get_latlong(post_code)
    
    #append the latitude and longitude data onto their respective list
    latitude.append(lat)
    longitude.append(long)
    
# create two new columns in the tor_df dataframe from the latitude and longitude lists
tor_df['Latitude'] = latitude
tor_df['Longitude'] = longitude
tor_df
```

This function will call on the Foursquare API to explore venues based on a location.

Inputs: 
- names: series containing neighborhood names
- latitudes: series containing latitude values
- longitudes: series containing longitude values
- radius: sets radial distance to explore around the latitude, longitude. default is 500 meters

Returns a Pandas dataframe nearby_venues containing venue information such as name, latitude, longitude, category


```python
# function that uses Foursquare API that get venues for specific locations
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[] # initialize a blank list to hold the venues
    for name, lat, lng in zip(names, latitudes, longitudes):
                    
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    # converts the nearby_venues list into a Pandas dataframe and renames the columns
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Create the dataframe of venues in the Toronto neighborhoods. We check the first 10 rows of the dataframe.

```python
# create separate series for the latitude, longitude, and neighborhoods from the tor_df_short dataframe
neighborhood_latitudes = tor_df_short['Latitude'] # neighborhood latitude value
neighborhood_longitudes = tor_df_short['Longitude'] # neighborhood longitude value
neighborhood_names = tor_df_short['Neighborhood'] # neighborhood names

# create a dataframe of Toronto venues by calling getNearbyVenues function
toronto_venues = getNearbyVenues(neighborhood_names, neighborhood_latitudes, neighborhood_longitudes)
toronto_venues.head(10)
```

This code creates a dataframe showing the frequency of each venue category within each neighborhood. This is done by first converting the categorical variable 'Venue Category' into an indicator variable. The frequency in each neighborhood is then calculated by grouping by the Neighborhood and taking the mean.

```python
# one hot encoding using get_dummies to convert categorical variables into indicator variables
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

# group by neighborhood and calculate the mean frequency of each venue type
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()

#check the first 10 rows
toronto_grouped.head(10)
```

Function which returns the most common venues in a neighborhood.

Inputs:
- row: dataframe row corresponding to a neighborhood
- num_top_venues: int number of most common venues to return

Returns a sorted dataframe of num_top_venues most common venues

```python
# function which returns the most common venues in a neighborhood
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Creates a dataframe with the top 10 venues for each neighborhood. The dataframe is sorted by neighborhood alphabetically.

```python
# variable to indicate how many top venues to retrieve for each neighborhood
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe to store the top venues in each neighborhood
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

# populate the dataframe with the top venues in each neighborhood by calling the return_most_common_venues function
for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

Using the venue frequency dataframe, this code clusters the neighborhoods into five clusters using k-means clustering.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
# n_init set to 15 to run the kmeans algorithm 20 times
kmeans = KMeans(n_clusters=kclusters, random_state=0, n_init=20).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

The cluster labels created from the k-means clustering is added to the sorted neighborhood and venues dataframe.

```python
# add clustering labels to the dataframe
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = tor_df_short

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head(10) # check the last columns!
```

```python
# reset the indexes of the dataframe
toronto_merged.reset_index(inplace=True, drop=True)
```

The clustered neighborhoods are now mapped using the Folium map library. Each neighborhood is indicated with a colored marker at its latitude, longitude coordinates in the color corresponding to its cluster.

```python
# get first latitude and longitude coordinates to use for creating the folium map of Toronto
latitude = toronto_merged['Latitude'][1]
longitude = toronto_merged['Longitude'][1]

# create map centered on Toronto
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
# iterate through the dataframe and add each neighborhood to the map wit the appropriate labels and colored marker
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(html = str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
